```python
from PIL import Image
import matplotlib.pyplot as plt

from diffae import DiffusionAutoEncodersInterface
```

```python
args = {
    'output': '../output/celebahq128',
    'model_ckpt': 'last_ckpt.pth',
}
mode = 'infer'
diffae = DiffusionAutoEncodersInterface(args, mode)
```

```python
image = Image.open('../assets/celebahq_sample/5.jpg')
plt.imshow(image)
```

```python
result = diffae.infer(image)

fig, ax = plt.subplots(1, 2, tight_layout=True)
for i, key in enumerate(['input', 'output']):
    ax[i].imshow(result[key])
    ax[i].set_title(key)
    ax[i].axis('off')
fig.show()
```
